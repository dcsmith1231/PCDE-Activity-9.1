```python
import csv
import pandas as pd
import mysql.connector as msql
from mysql.connector import Error
```

```python
spenddata = pd.read_csv('spend.csv', index_col=False, delimiter = ',')
df = pd.read_excel('mrtssales92-present.xls', sheet_name = "2020",skiprows=4, skipfooter=47)
df.head(5)
```

```python
try:
    conn = msql.connect(host='localhost', database='spend', user='root', password = 'mit123')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS `year2020`;')
        print('Creating table...')
        cursor.execute("CREATE TABLE year2020 (Kind_of_Business varchar(255), Jan varchar(255),Feb varchar(255),Mar varchar(255),Apr varchar(255),May varchar(255),Jun varchar(255),Jul varchar(255), Aug varchar(255),Sep varchar(255),Oct varchar(255),Nov varchar(255), Dece varchar(255))")
        print('Table is created....')
        #loop through dataframe
        for i, row in spenddata.iterrows():
            sql = "INSERT INTO spend.year2020 VALUES (%s,%s, %s,%s, %s,%s, %s,%s, %s,%s, %s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we have to commit to save our changes
            conn.commit()
except Error as e:
    print("error while connecting to MySQL", e)
```
